## Edge Detection

```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
import os
img1 = '1.jpg'
img2 = '2.jpg'
img3 = '3.jpg'
img4 = '4.jpg'
img5 = '5jpg'
img6 = '6jpg'
img7 = '7.jpg'
img8 = '8.jpg'
img9 = '9.jpg'
img10 = '10.jpg'
img11 = 'plat.jpg'

path = img11             # UBAH PATH DISINI
path2 = img8
path3 = img11
filename = os.path.splitext(path)[0]
img = cv2.imread(path,0)
img.shape
```

### 1. Sobel Operator

```python
sobelx = cv2.Sobel(img,-1,1,0,ksize=5) # Image , DDepth = -1 (uint8) would be the result
                                        # Caution : you may need to try cv2.CV_64F for negative values to be considered
                                        # Black to white and White to Black Both would appear.
sobely = cv2.Sobel(img,-1,0,1,ksize=5) # 1 , 0  for dx   and  0 , 1 for dy

plt.figure(figsize=(16, 12))

plt.subplot(1,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
output_sobel = 'output/'+filename+'_sobel.jpg'
plt.savefig(output_sobel,bbox_inches = 'tight', pad_inches = 0.1)
plt.show()
```

### 2. Scharr Filter

```python
# img = cv2.imread('E://Computer Vision//Data//sudoko.jpg',0)
img = cv2.imread(path,0)

scharrx = cv2.Scharr(img, -1,1,0,5)  # img ,ddepth, dx , dy, kernel 
scharry = cv2.Scharr(img, -1,0,1,5) 
plt.figure(figsize=(16, 12))

plt.subplot(1,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,2),plt.imshow(scharrx,cmap = 'gray')
plt.title('Scharr X'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(scharry,cmap = 'gray')
plt.title('Scharr Y'), plt.xticks([]), plt.yticks([])
output_scharr = 'output/'+filename+'_scharr.jpg'
plt.savefig(output_scharr,bbox_inches = 'tight', pad_inches = 0.1)
plt.show()
```

### 3. Laplacian of Gradient 

```python
# img = cv2.imread('E://Computer Vision//Data//sudoko.jpg',0)
img = cv2.imread(path,0)
laplacian = cv2.Laplacian(img, -1,ksize=3)  # img , ddepth , kernel
plt.figure(figsize=(16, 12))

plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(laplacian,cmap = 'viridis')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
output_laplacian = 'output/'+filename+'_laplacian.jpg'
plt.savefig(output_laplacian,bbox_inches = 'tight', pad_inches = 0.1)
plt.show()
```

### 4. Canny Edge Detector

```python
# img = cv2.imread('E://Computer Vision//Data//sudoko.jpg',0)
img = cv2.imread(path,0)
edges = cv2.Canny(img,160,190) # image , min threshold , max threshold
plt.figure(figsize=(16, 12))

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.suptitle('')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
# plt.subplot(124),plt.imshow(edges,cmap = 'gray')
# plt.title('Edge Image Colored'), plt.xticks([]), plt.yticks([])
output_canny = 'output/'+filename+'_canny.jpg'
plt.savefig(output_canny,bbox_inches = 'tight', pad_inches = 0.1)
plt.show()
```

```python
# img = cv2.imread('E://Computer Vision//Data//car.jpg',0)
img = cv2.imread(path,0)
edges = cv2.Canny(img,100,200)
plt.figure(figsize=(16, 12))

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
